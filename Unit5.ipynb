### 데이터 읽어오기

서울의 기온 데이터를 읽어와서 한 줄씩 출력하겠다.

이제 데이터의 맨 윗줄에 있는 헤더 부분을 `next()` 함수를 사용해 제외시킨 후, 최고 기온 데이터만 출력해보자.

```python
f = open('preprocessed_data_v1.csv', encoding='cp949')
data = csv.reader(f)
next(data)

for row in data:
    print(row[-1])
```

### 데이터 리스트에 저장하기
---
최고 기온 데이터를 저장해보자. 날짜 순으로 저장하고자 `result` 리스트를 새로 만들고 최고 기온 데이터가 존재하는 날의 최고 기온 데이터를 `result` 리스트에 저장하자.

```python
f = open('preprocessed_data_v1.csv', encoding='cp949')
data = csv.reader(f)
next(data)

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
    if row[-1] != '' : #최고 기온 데이터 값이 존재한다면
        result.append(float(row[-1])) #result 리스트에 최고 기온 값 추가
        
print(result)
```

데이터가 너무 많아서 보기 힘들다 `len()` 함수로 데이터의 갯수를 확인하자.

```python
len(result)
```

최고 기온 데이터를 꺾은선 그래프로 표현하자.

```python
import matplotlib.pyplot as plt

plt.plot(result, 'r') #result 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.show() #그래프 나타내기
```

## 3. 날짜 데이터 추출하기
---
파이썬이 제공하는 문자열 함수 중 `split()`는 사용자가 설정하는 특정 문자를 기준으로 문자열을 분리한다.

```python
# split() 예시 코드
s = 'hello python'
print(s.split())
```

이제 `split()` 함수를 사용해 1년 중 여름의 정점인 8월의 최고 기온 데이터만 추출해서 그래프로 그려보자.

```python
f = open('preprocessed_data_v1.csv', encoding='cp949')
data = csv.reader(f)
next(data)

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
    if row[-1] != '' : #최고 기온 데이터 값이 존재한다면
        result.append(float(row[-1])) #result 리스트에 최고 기온 값 추가
        
print(result)
```

```python
f = open('preprocessed_data_v1.csv', encoding='cp949')
data = csv.reader(f)
next(data)

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
    if row[-1] != '' : #최고 기온 값이 존재한다면
        if row[0].split('-')[1] == '08' : #8월에 해당하는 값이라면
            result.append(float(row[-1])) # result 리스트에 최고 기온 값 추가
            
plt.plot(result, 'hotpink') # result 리스트에 저장된 값을 hotpink 색으로 그리기
plt.show() # 그래프 나타내기
```

8월의 데이터만 모아서 그래프로 나타내니 훨씬 간결하다. 이번에는 매년 2월 14일을 기준으로 그래프를 그려보자.

```python
f = open('preprocessed_data_v1.csv', encoding='cp949')
data = csv.reader(f)
next(data)

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
    if row[-1] != '' : #최고 기온 값이 존재한다면
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14' : #2월 14일에 해당하는 값이라면
            result.append(float(row[-1])) # result 리스트에 최고 기온 값 추가
            
plt.plot(result, 'hotpink') # result 리스트에 저장된 값을 hotpink 색으로 그리기
plt.show() # 그래프 나타내기
```

지난 100년간 2월 14일의 온도변화를 알 수 있다. 최근들어 기온이 높아진 것을 확인 할 수 있다.

좀 더 자세히 보기위해 1983년 이후 데이터만 추출해서 그래프로 그려보자. 최고 기온뿐만 아니라 최저 기온 데이터도 함꼐 나타내보자.

```python
f = open('preprocessed_data_v1.csv', encoding='cp949')
data = csv.reader(f)
next(data)

high = [] # 최고 기온 값을 저장할 리스트 high 생성
low = [] # 최저 기온 값을 저장할 리스트 low 생성

for row in data:
    if row[-1] != '' : #최고 기온 값과 최저 기온 값이 존재한다면
        if 1983 <= int(row[0].split('-')[0]): # 1983년 이후 데이터 라면
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14' : #2월 14일이라면
                high.append(float(row[-1])) # 최고 기온 값을 high 리스트에 저장
                low.append(float(row[-2])) # 최저 기온 값을 low 리스트에 저장
                
plt.plot(high, 'hotpink') # high 리스트에 저장된 값을 hotpink 색으로 그리기
plt.plot(low, 'skyblue') # low 리스트에 저장된 값을 skyblue 색으로 그리기
plt.show() # 그래프 나타내기
```
