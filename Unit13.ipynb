## 1. numpy 라이브러리
---

```python
import matplotlib.pyplot as plt
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)
plt.figure(dpi = 300)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

코드 2행을 보니 `numpy` 라이브러리를 `np`라는 별명으로 부르는 것을 확인할 수 있다. 이 코드는 파이썬 리스트를 사용해서 다음과 같이 구현할 수 있다.

```python
t = []
p2 = []
p3 = []

for i in range(0, 50, 2) :
    t.append(i / 10)
    p2.append((i / 10) ** 2)
    p3.append((i / 10) ** 3)
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

`numpy` 라이브러리를 사용한 코드와 파이썬 리스트를 사용한 코드의 실행 결과는 똑같지만, `numpy` 라이브러리를 사용한 경우가 훨씬 적은 수의 코드로 작성하여 간결하다. 그리고 홈페이지에 소개된 내용을 더 살펴보면 이 예제뿐만 아니라 다른 예제에서도 `np`로 시작하는 `numpy` 라이브러리를 활용하는 코드들을 많이 볼 수 있다.

실행 결과를 확인해보니 리스트와 비슷하게 생긴 `numpy.ndarray` 타입의 데이터가 생성된 것을 볼 수 있다. `ndarray`에서 `nd`는 `N–Dimensional`, 즉 `N` '차원'이라는 의미이고 `array`는 '배열'이라는 의미한다

`choice()` 함수는 `random` 라이브러리에서는 표현하기 어려운 랜덤 데이터를 생성한다.

```python
print(np.random.choice(6, 10))
```

실행 결과에서 짐작할 수 있듯이 0-5 사이의 숫자를 랜덤하게 10번 선택했다. 만약 한 번 뽑은 숫자를 다시 뽑지 못하게 하고 싶다면 `replace` 속성을 `False`로 설정하면 된다.

```python
print(np.random.choice(10, 6, replace = False))
```

여기에 확률을 지정할 수 있는 `p` 속성을 추가하겠다. `p` 속성은 각 경우의 수가 발생할 확률을 정할 수 있다. 0-5까지의 경우가 발생할 수 있으므로 0이 나올 확률부터 5가 나올 확률까지 하나씩 지정할 수 있다. 따라서 이 경우에는 `p` 속성에 6개의 확률이 있어야 하고 그 합이 반드시 1이어야 한다. 0은 0.1, 1은 0.2, 2는 0.3, 3은 0.2, 4는 0.1, 5는 0.1의 확률을 지정하여 프로그램을 실행한다.

```python
print(np.random.choice(6, 10, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1]))
```

## 4. numpy array 생성하기
---
`numpy` 라이브러리에서는 `ndarray`라는 특별한 데이터 타입의 배열이 사용된다.

```python
a = np.array([1,2,3,4])
print(a)
```

실행 결과를 살펴보면 양 끝에 `[]`가 붙는 점이 리스트와 비슷하다. 하지만 `numpy array`는 리스트 내 요소를 구분하는 쉼표(`,`)가 없다는 차이점이 있다. `numpy array`는 생긴 모습도 리스트와 비슷하지만, 리스트의 특징인 인덱싱과 슬라이싱을 할 수 있다는 점도 비슷하다.

```python
print(a[1], a[-1]) # a의 1번 인덱스 값, -1번 인덱스 값 출력
print(a[1:])  # a의 1번 인덱스를 기준으로 슬라이싱 결과 출력
```

정수, 문자열, 리스트 등 다양한 데이터 타입을 담을 수 있었던 리스트와는 달리 `numpy array`에는 한 가지 타입의 데이터만을 저장할 수 있다.

다음과 같이 3개의 정수와 1개의 문자가 저장된 리스트를 배열로 변환하면 모두 같은 문자 타입으로 변환된다.

```python
a = np.array([1,2,'3',4])
print(a)
```

`zeros()`, `ones()`, `eye()` 같은 다양한 함수를 사용해서 배열을 초기화할 수도 있다.

```python
a = np.zeros(10)  # 0으로 이루어진 크기가 10인 배열 생성
print(a)
```

```python
a = np.ones(10)   # 1로 이루어진 크기가 10인 배열 생성
print(a)
```

```python
a = np.eye(3)  # 3행 x 3열의 단위 배열 생성
print(a)
```

앞서 살펴본 세 가지 예는 0 또는 1로 배열을 채우고 싶을 때 무척 유용하다. 하지만 0과 1이 아닌 연속된 숫자로 데이터를 생성하는 것은 어떻게 할 수 있을까? 

다음은 연속된 숫자로 데이터를 생성하는 것을 보여주는 예시로, `numpy`의 `arange()` 함수는 값을 3개까지 입력하여 사용할 수 있다.

```python
print(np.arange(3))     # arange( ) 함수에 1개 값 입력
print(np.arange(3,7))   # arange( ) 함수에 2개 값 입력
print(np.arange(3,7,2)) # arange( ) 함수에 3개 값 입력
```

`arange()` 함수에 값을 하나만 입력할 경우, 0부터 해당 숫자보다 1만큼 작은 정수까지 저장된 배열이 만들어진다. 값을 두 개 입력할 경우, 첫 번째 숫자부터 두 번째 숫자보다 1만큼 작은 정수까지 저장된 배열이 만들어진다. 값을 세 개 입력할 경우, 첫 번째 숫자부터 두 번째 숫자보다 1만큼 작은 정수까지의 범위에서 세 번째 숫자만큼의 간격을 둔 숫자가 저장된 배열이 만들어진다.

다른 숫자를 생성하는 방법으로 `numpy`의 `linspace()` 함수를 사용할 수 있다.

```python
a = np.arange(1, 2, 0.1)   # 1이상 2미만 구간에서 0.1 간격으로 실수 생성
b = np.linspace(1, 2, 11)  # 1부터 2까지 11개 구간으로 나눈 실수 생성
print(a)
print(b)
```

`numpy`의 `linspace()` 함수는 특정 구간을 쪼개어 값을 생성한다는 점에서 `arange()` 함수와 비슷하다. 하지만 `arange()` 함수에서는 간격을 지정하고 `linspace()`함수에서는 구간의 개수를 지정한다는 점에서 차이가 있다.

특정 간격에 해당하는 값을 생성하고 싶을 때는 `arange()` 함수를, 특정 개수의 구간으로 나눈 값을 생성하고 싶을 때는 `linspace()` 함수를 사용할 수 있다.

```python
a = np.arange(-np.pi, np.pi, np.pi/10)
b = np.linspace(-np.pi, np.pi, 20)
print(a)
print(b)
```

## 5. numpy array의 다양한 활용
---
그런데 `np.zeros()`, `np.ones()` 함수는 있는데 왜 `np.twos()`, `np.threes()` 함수는 없을까? 만약 초깃값이 5인 배열을 100개 만들고 싶으면 어떻게 하면 될까? 바로 이 점이 `numpy array`를 우리가 알아야 하는 가장 큰 이유인 것 같다.

```python
a = np.zeros(10) + 5
print(a)
```

이번에는 배열에 연산이 아닌 함수를 적용하자.

```python
a = np.linspace(1, 2, 11)
print(np.sqrt(a))  # a값의 제곱근을 출력함
```

배열에 어떤 연산이나 함수를 적용하면 배열의 모든 값이 한꺼번에 계산되는 것을 알 수 있다. 이런 특성을 활용하면 다음과 같이 아주 간단한 코드로 그래프를 나타낼 수 있다.

```python
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.show()
```

이 코드의 3행은 `pi` 값을 기준으로 특정 범위를 생성하고, 범위 내 숫자들을 `pi` 값을 100으로 나눈 값만큼 간격을 두어 배열 `a`에 저장한 것이다. 그런 다음 `sin()` 함수를 `a`에 저장된 값 전체에 적용한 후 그래프로 표현하였다. 즉, 특정 구간의 `sin()` 함수의 모습을 쉽게 시각화할 수 있다.

이를 응용하여, 다음은 배열 `a`에 저장된 값을 바탕으로 `sin()` 함수와 `cos()` 함수를 표현한 것이다. 6행을 보면 `sin()` 함수의 평행이동도 쉽게 구현할 수 있음을 보여준다.



```python
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
plt.plot(a+np.pi/2, np.sin(a))
plt.show()
```

`마스크(mask)` 기능에 대해 알아보겠다. 마스크는 어떤 조건에 부합하는 데이터만 선별적으로 저장하기 위한 기능이다.

예를 들어 다음처럼 `a`라는 배열을 만들었을 경우 0보다 작은 수는 -5부터 –1까지이다.

```python
a = np.arange(-5, 5)
print(a)
```

`numpy`에서는 이런 상황을 배열 전체에 조건을 적용해서 다음처럼 표현할 수 있다. 그러면 배열에서 조건에 부합하는 데이터는 `True`, 그렇지 않은 데이터는 `False`인 마스크가 생성된다.

```python
print(a<0)
```

이 마스크를 다시 배열에 적용하면 다음과 같이 조건에 부합하는 데이터만 남는다.

```python
print(a[a<0])
```

그리고 마스크를 다음과 같이 변수에 넣어서 사용할 수도 있다.

```python
mask1 = abs(a) > 3
print(a[mask1])
```

또 이렇게 몇 개의 마스크를 연결해서 사용할 수도 있다.

```python
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0
print(a[mask1+mask2])   # 둘 중 하나의 조건이라도 참일 경우
print(a[mask1*mask2])   # 두 가지 조건이 모두 참일 경우
```

`numpy` 라이브러리를 사용하여 재미있는 버블 차트 그리기

```python
x = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
size = np.random.rand(1000) *100
mask1 = abs(x) >50 # x에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
mask2 = abs(y) >50 # y에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
x = x[mask1 + mask2] # mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1 + mask2] # mask1과 mask2 중 하나라도 만족하는 값 저장

size = np.random.rand(x.shape[0]) * 100

plt.scatter(x, y, s = size, c = x, cmap = 'jet', alpha = 0.7)
plt.colorbar()
plt.show()
```
